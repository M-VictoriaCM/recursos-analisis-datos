# DATOS DATAFRAME

Analizar como se componen cada dataframe, saber cuál es su tamaño, ver que tipo de datos tiene ademas de conocer la cantidad de nulos.

## Función datos DataFrames

```python
#Definimos una función para obtener información importante sobre la cantidad y tipos de datos
def datosDataFrames(df):
    # Extraer información relevante
    column_names = df.columns.tolist()
    data_types = df.dtypes.tolist()

    # Crear tabla PrettyTable
    table = PrettyTable()

    # Información del DataFrame
    table.add_column("Característica", ["Tamaño"])
    table.add_column("Valor", [df.shape])
    table.title = "Información del DataFrame"
    print(table)

    # Análisis de tipos de datos

    table = PrettyTable()
    table.title = "Análisis de tipos de datos(Número de columnas:",df.shape[1]
    table.add_column("Atributos",column_names)
    table.add_column("Tipo de dato", df.dtypes)
    table.add_column("Non-null",df.notnull().sum())
    print(table)

    # Análisis de valores nulos
    table = PrettyTable()
    table.add_column("Columna", df.columns)
    table.add_column("Valores nulos", df.isnull().sum())
    table.title = "Análisis de valores nulos"
    print(table)

    # Análisis de duplicados
    table = PrettyTable()
    table.add_column("Cantidad", [df.duplicated().sum()])
    table.title = "Análisis de duplicados"
    print(table)


```

# TRANSFORMACIÓN DATAFRAMES

Para evitar futuros problemas se debe

asegurarse que los datos sean homogeneos.

## Función TransformarFecha

```python
#Definimos nuestra fucnion transformarFecha
#Los parametros son el dataframe y la lista con las columnas a transformar
def transformarFechas(df, columnas_fechas): #Los parametros son el dataframe y la lista con las columnas a transformar
    df[columnas_fechas] = df[columnas_fechas].apply(pd.to_datetime) #Transforma las columnas de columnas_fechas a datetime
    return df
```

## Función transformarString

```python
def transformarString(df, columnas_string):
  for columna in columnas_string:
    df[columna] =df[columna].astype("string")
  return df
```

## Función convertirMayusculas

```python
def convertirMayusculas(df, columna):
  df[columna] = df[columna].str.upper()
  return df
```

## Función convertirMinusculas

```python
def convertirMinusculas(df, columna):
  df[columna] =df[columna].str.lower()
  return df
```

# ANALISÍS DATAFRAMES

Después de transformar los DataFrames podemos realizar un analisís de los datos que contienen, para poder extraer información relevante( **insights** )

## Función analisisDataFrames

```python
#Definimos una función para obtener insights relevantes de nuestros dataframes
def analisisDataFrames(df): #Como parametro tenemos los dataframes
  print("----------")
  print("Análisis variables cuantitativas")
  columnas_numericas = df.select_dtypes(include=['int', 'float']).columns #Extraemos las columnas numericas
  if not columnas_numericas.empty: #Aplicamos un if en caso de que no tenga columnas numéricas el dataframe
    print(df[columnas_numericas].describe()) #Usamos el describe para aquellas columnas cuantitativas
  print("----------")
  print("----------")
  print("Análisis variables cualitativas")
  columnas_string = df.select_dtypes(include=['string']).columns #Extraemos las columnas string
  if not columnas_string.empty: #Aplicamos un if en caso de que no tenga columnas del tipo string el dataframe
    print(df[columnas_string].describe()) #Usamos el describe para aquellas columnas cualitativas
```
